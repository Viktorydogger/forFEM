```python
from tkinter import *

def check_pass():
    if entry_p.get() and entry_m.get():
        label_welcome.configure(text="Сейчас построим сетку")
    else:
        label_welcome.configure(text="данные не могут быть обработаны")
    m = entry_m.get()
    m1 = entry_m1.get()
    p = entry_p.get()
    N = entry_N.get()
    r = entry_r.get()
    d1 = entry_d1.get()
    d2 = entry_d2.get()
    

root = Tk()
root.title("Введите параметры сетки")
label_d1 = Label(root, text="Длина горизонтальной грани")
label_d1.pack()
entry_d1 = Entry(root, width=10)
entry_d1.pack()
label_d2 = Label(root, text="Длина вертекальной грани")
label_d2.pack()
entry_d2 = Entry(root, width=10)
entry_d2.pack()
label_r = Label(root, text="Радиус отверстия")
label_r.pack()
entry_r = Entry(root, width=10)
entry_r.pack()
label_p = Label(root, text="p")
label_p.pack()
entry_p = Entry(root, width=10)
entry_p.pack()
label_m = Label(root, text="m")
label_m.pack()
entry_m = Entry(root, width=10)
entry_m.pack()
label_m1 = Label(root, text="m1")
label_m1.pack()
entry_m1 = Entry(root, width=10)
entry_m1.pack()
label_N = Label(root, text="количество итераций")
label_N.pack()
entry_N = Entry(root, width=10)
entry_N.pack()
button = Button(root, text="enter", command=check_pass)
button.pack()
label_welcome = Label(root)
label_welcome.pack()
root.mainloop()
```
